```python
def functionFirstPage(linkName):
  r = requests.get(linkName)
  soup = BeautifulSoup(r.text, 'html.parser')
  q1 = soup.find('div', {"class": 'Column-main test-drives'})
  for news in q1.findAll('h2', {'class':'Post-title'}):
    headerArray.append(news.text)

def functionToNewPage(number, linkName):
  req = requests.get(linkName + str(number))
  soup = BeautifulSoup(req.text, 'html.parser')
  q1 = soup.find('div', {"class": 'Column-main test-drives'})
  for news in q1.findAll('h2', {'class':'Post-title'}):
    headerArray.append(news.text)

def toBodyLink(link):
  r = requests.get(link)
  soup = BeautifulSoup(r.text, 'html.parser')
  q1 = soup.find('div', {"class": 'Column-main test-drives'})
  for news in q1.findAll('div', {"class":"Post-info"}):
      link = news.find('a', {'class':'Post-link'})
      links.append(link['href'])

def toBody():
  body = []
  for link in links:
    page = requests.get(link, allow_redirects=False)
    soup = BeautifulSoup(page.text, 'html.parser')
    bodyS = soup.find('div', {'class':'Article-bodyText'})
    if bodyS is None:
      print('Empty body')
    else:
      for p in bodyS.findAll('p'):
        body.append(p.text)
      title = soup.find('h1', {'class':'u-entryTitle'})
      hArray.append(title.text)
    bodyArray.append(''.join(body))
    body = []
```

```python
def concatenate_text(examples):
  if examples['Heading'] is None or examples['Body'] is None:
    return {"Text": " "}
  else:
    return {"Text": examples['Heading'] + " \n " + examples['Body']}
dataset = dataset.map(concatenate_text)
```

```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained('ddobokki/electra-small-nli-sts')
model = AutoModel.from_pretrained('ddobokki/electra-small-nli-sts', from_pretrained = True)
```

```python
# ELECTRA pooling
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

def get_embeddings(text_list):
    encoded_input = tokenizer(text_list, padding=True, truncation=True, return_tensors="pt", max_length = 256)
    encoded_input = {k: v for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return mean_pooling(model_output, encoded_input['attention_mask'])
```

```python
embedding = get_embeddings(dataset["Text"][0])
embedding.shape
```

```python
with torch.no_grad():
  embeddings_dataset = dataset.map(lambda x: {"embeddings": get_embeddings(x["Text"]).numpy()[0]})
```

```python
import faiss.contrib.torch_utils
question = 'NASA'
with torch.no_grad():
  question_embedding = get_embeddings([question]).numpy()
question_embedding.shape
```

```python
scores, samples = embeddings_dataset.get_nearest_examples('embeddings', question_embedding, k=5)
samples_df = pd.DataFrame.from_dict(samples)
samples_df["scores"] = scores
samples_df.sort_values("scores", ascending=True, inplace=True)


for _, row in samples_df.iterrows():
    print(f"HEADING: {row.Heading}")
    print(f"BODY: {row.Body}")
    print(f"SCORE: {row.scores}")
    print("=" * 50)
    print()
```
